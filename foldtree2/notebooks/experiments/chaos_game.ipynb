```python
import numpy as np

def ngon_vertices(N, phase=0.0):
    angles = np.linspace(0, 2*np.pi, N, endpoint=False) + phase
    return np.stack([np.cos(angles), np.sin(angles)], axis=1)

def cgr_scale_factor(N):
    m = int(np.floor(N / 4))
    num = np.sin(np.pi / N)
    den = num + np.sin(np.pi / N + 2*np.pi*m / N)
    return 1.0 - num / den

def cgr_sequence(tokens, N, alpha=None, phase=0.0, x0=None):
    verts = ngon_vertices(N, phase=phase)
    if alpha is None:
        alpha = cgr_scale_factor(N)
    if x0 is None:
        x = np.zeros(2, dtype=float)
    else:
        x = np.array(x0, dtype=float)

    pts = np.zeros((len(tokens), 2), dtype=float)
    for i, tok in enumerate(tokens):
        x = (1 - alpha) * x + alpha * verts[tok]
        pts[i] = x
    return pts, verts, alpha

def fcgr(points, bins=64, lim=1.05):
    H, xe, ye = np.histogram2d(
        points[:,0], points[:,1],
        bins=bins,
        range=[[-lim, lim], [-lim, lim]]
    )
    return H.T
```
